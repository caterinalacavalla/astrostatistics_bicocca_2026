```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
from scipy.stats import poisson
```

```python
def poiss(x, mu):
    poiss_values = []
    for i in range (len(x)):
        poiss_values.append(scipy.stats.poisson.pmf(x[i], mu))
    return poiss_values
    
```

```python
#data

deaths = []
for i in range(109):
    deaths.append(0)
for i in range (65):
    deaths.append(1)
for i in range (22):
    deaths.append(2)
for i in range (3):
    deaths.append(3)
for i in range (1):
    deaths.append(4)


mu = np.mean(deaths)
x = np.arange(0, 5, 0.001)
poi = poiss(x, mu)
print (x, poi)
plt.figure(figsize = (10,6))
plt.hist(deaths, 5, histtype = 'bar', density=True)
#plt.plot(poisson.pdf(mean), 'o', color = "thistle", label = "mean")
plt.scatter(x, poi, color = "palegreen", label = "p_mean")
plt.legend()
plt.xlabel('Number of deaths')
plt.ylabel('Number of groups')
plt.title('Distribution of deaths in cavalry corps')
plt.show()



```
