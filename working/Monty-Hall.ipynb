```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
#simulo 3 porte , in una c'è una macchina, in due c'è una capra
N = np.arange(10, 10010, 500)

switcher_tot = []
conservative_tot = []
newcomer_tot = []



for j in N:
    switcher_wins = 0
    conservative_wins = 0
    newcomer_wins = 0

    for i in range(0, j):
        doors = np.zeros(3)
        doors[random.randint(0, 2)] = 1  

        #the switcher and conservative
        choice = random.randint(0,2)
        #probability = 1/3
        opendoor = random.randint(0,2)
        while opendoor == choice or doors[opendoor] == 1:
            opendoor = random.randint(0,2)

        secondchoice = random.randint(0,2)
        while secondchoice == opendoor or secondchoice == choice:
            secondchoice = random.randint(0,2)
        #probability = 2/3

        if doors[secondchoice] == 1:
            switcher_wins += 1
        if doors[choice] == 1:
            conservative_wins += 1

        #the newcomer
        closed_doors = [door for door in range(len(doors)) if door != opendoor]
        newcomer_choice = random.choice(closed_doors)
        #probability = 1/2

        if doors[newcomer_choice] == 1:
            newcomer_wins += 1

    switcher_tot.append(switcher_wins)
    conservative_tot.append(conservative_wins)
    newcomer_tot.append(newcomer_wins)

twothirdsline  = 2/3 * N
onehalfline = 1/2 *N 
onethirdline = 1/3 *N
plt.figure (figsize = (10, 6))
plt.plot (N, twothirdsline, color = "slategray", linestyle = "--", label = "P = 2/3")
plt.plot (N, switcher_tot, color = "slategray", label = "switcher wins")
plt.plot (N, onethirdline, color = "thistle", linestyle = "--", label = "P = 1/3")
plt.plot (N, conservative_tot, color = "thistle", label = "conservative wins")
plt.plot (N, onehalfline, color = "palegreen", linestyle = "--", label = "P = 1/2")
plt.plot (N, newcomer_tot, color = "palegreen", label = "newcomer wins")
plt.xlabel ("Number of tries")
plt.ylabel ("Number of wins")
plt.title ("Monty Hall")
plt.legend()
plt.tight_layout()
plt.show()




#più da vicino
N = np.arange(1, 100, 1)
switcher_tot = []
conservative_tot = []
newcomer_tot = []

for j in N:
    switcher_wins = 0
    conservative_wins = 0
    newcomer_wins = 0

    for i in range(0, j):
        doors = np.zeros(3)
        doors[random.randint(0, 2)] = 1  

        #the switcher and conservative
        choice = random.randint(0,2)
        #probability = 1/3
        opendoor = random.randint(0,2)
        while opendoor == choice or doors[opendoor] == 1:
            opendoor = random.randint(0,2)

        secondchoice = random.randint(0,2)
        while secondchoice == opendoor or secondchoice == choice:
            secondchoice = random.randint(0,2)
        #probability = 2/3

        if doors[secondchoice] == 1:
            switcher_wins += 1
        if doors[choice] == 1:
            conservative_wins += 1

        #the newcomer
        closed_doors = [door for door in range(len(doors)) if door != opendoor]
        newcomer_choice = random.choice(closed_doors)
        #probability = 1/2

        if doors[newcomer_choice] == 1:
            newcomer_wins += 1

    switcher_tot.append(switcher_wins)
    conservative_tot.append(conservative_wins)
    newcomer_tot.append(newcomer_wins)

twothirdsline  = 2/3 * N
onehalfline = 1/2 *N 
onethirdline = 1/3 *N
plt.figure (figsize = (10, 6))
plt.plot (N, twothirdsline, color = "slategray", linestyle = "--", label = "P = 2/3")
plt.plot (N, switcher_tot, color = "slategray", label = "switcher wins")
plt.plot (N, onethirdline, color = "thistle", linestyle = "--", label = "P = 1/3")
plt.plot (N, conservative_tot, color = "thistle", label = "conservative wins")
plt.plot (N, onehalfline, color = "palegreen", linestyle = "--", label = "P = 1/2")
plt.plot (N, newcomer_tot, color = "palegreen", label = "newcomer wins")
plt.xlabel ("Number of tries")
plt.ylabel ("Number of wins")
plt.title ("Monty Hall, a closer look")
plt.legend()
plt.tight_layout()
plt.show()


#invece del closer look, faccio 100 tries tante volte e metto in un'istogramma i valori di probabilità ottenuti

N = 100
tries = 100
num_experiments = 1000

switcher_probs = []
conservative_probs = []
newcomer_probs = []

for _ in range(num_experiments):
    switcher_wins = 0
    conservative_wins = 0
    newcomer_wins = 0

    for i in range(tries):
        doors = np.zeros(3)
        doors[random.randint(0, 2)] = 1  

        #the switcher and conservative
        choice = random.randint(0,2)
        #probability = 1/3
        opendoor = random.randint(0,2)
        while opendoor == choice or doors[opendoor] == 1:
            opendoor = random.randint(0,2)

        secondchoice = random.randint(0,2)
        while secondchoice == opendoor or secondchoice == choice:
            secondchoice = random.randint(0,2)
        #probability = 2/3

        if doors[secondchoice] == 1:
            switcher_wins += 1
        if doors[choice] == 1:
            conservative_wins += 1

        #the newcomer
        closed_doors = [door for door in range(len(doors)) if door != opendoor]
        newcomer_choice = random.choice(closed_doors)
        #probability = 1/2

        if doors[newcomer_choice] == 1:
            newcomer_wins += 1

    switcher_probs.append(switcher_wins / tries)
    conservative_probs.append(conservative_wins / tries)
    newcomer_probs.append(newcomer_wins / tries)

plt.figure(figsize=(10, 6))
plt.hist(switcher_probs, bins=20, alpha=0.5, label='Switcher', color='slategray')
plt.hist(conservative_probs, bins=20, alpha=0.5, label='Conservative', color='thistle')
plt.hist(newcomer_probs, bins=20, alpha=0.5, label='Newcomer', color='palegreen')
plt.xlabel("Probability of winning")
plt.ylabel("Frequency")
plt.title("Monty Hall, probability distribution")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Generalization to N doors  
#simulo N porte , in una c'è una macchina, nelle altre c'è una capra, il presentatore ne apre p<N

#------------------ number of opened doors fixed---------------------------
N_min = 3
N_max = 100 
doors_range = range(N_min, N_max+1)
tries = 1000

switcher_tot = []
conservative_tot = []
newcomer_tot = []

for N in doors_range:
    switcher_wins = 0
    conservative_wins = 0
    newcomer_wins = 0
    p = N-2

    for i in range(0, tries):

        doors = np.zeros(N)
        doors[random.randint(0, N-1)] = 1  

        #the conservative
        choice = random.randint(0,N-1)
        if doors[choice] == 1:
            conservative_wins += 1

        available_to_open = [d for d in range(N) if d != choice and doors[d] != 1]
        open_doors = random.sample(available_to_open, p)

        #the switcher
        available_for_switch = [d for d in range(N) if d not in open_doors and d != choice]
        secondchoice = random.choice(available_for_switch)
        if doors[secondchoice] == 1:
            switcher_wins += 1

        #the newcomer
        available_for_newcomer = [d for d in range(N) if d not in open_doors]
        newcomer_choice = random.choice(available_for_newcomer)
        if doors[newcomer_choice] == 1:
            newcomer_wins += 1

    switcher_tot.append(switcher_wins)
    conservative_tot.append(conservative_wins)
    newcomer_tot.append(newcomer_wins)

plt.figure (figsize = (10, 6))
plt.plot (doors_range, switcher_tot, color = "slategray", label = "switcher wins")
plt.plot (doors_range, conservative_tot, color = "thistle", label = "conservative wins")
plt.plot (doors_range, newcomer_tot, color = "palegreen", label = "newcomer wins")
plt.xlabel ("Number of doors")
plt.ylabel (f"Number of wins, out of {tries} tries")
plt.title ("Monty Hall vs number of doors, p = N-2")
plt.legend()
plt.tight_layout()
plt.show()

#------------------ total number of doors fixed---------------------------
N = 100
tries = 1000

p_range = range(1, N-2)

switcher_tot = []
conservative_tot = []
newcomer_tot = []

for p in p_range:
    switcher_wins = 0
    conservative_wins = 0
    newcomer_wins = 0

    for i in range(0, tries):

        doors = np.zeros(N)
        doors[random.randint(0, N-1)] = 1  

        #the conservative
        choice = random.randint(0,N-1)
        if doors[choice] == 1:
            conservative_wins += 1

        available_to_open = [d for d in range(N) if d != choice and doors[d] != 1]
        open_doors = random.sample(available_to_open, p)

        #the switcher
        available_for_switch = [d for d in range(N) if d not in open_doors and d != choice]
        secondchoice = random.choice(available_for_switch)
        if doors[secondchoice] == 1:
            switcher_wins += 1

        #the newcomer
        available_for_newcomer = [d for d in range(N) if d not in open_doors]
        newcomer_choice = random.choice(available_for_newcomer)
        if doors[newcomer_choice] == 1:
            newcomer_wins += 1

    switcher_tot.append(switcher_wins)
    conservative_tot.append(conservative_wins)
    newcomer_tot.append(newcomer_wins)

plt.figure (figsize = (10, 6))
plt.plot (p_range, switcher_tot, color = "slategray", label = "switcher wins")
plt.plot (p_range, conservative_tot, color = "thistle", label = "conservative wins")
plt.plot (p_range, newcomer_tot, color = "palegreen", label = "newcomer wins")
plt.xlabel ("Number of opened doors")
plt.ylabel (f"Number of wins, out of {tries} tries")
plt.title ("Monty Hall vs number of opened doors, N = 100")
plt.legend()
plt.tight_layout()
plt.show()

```
