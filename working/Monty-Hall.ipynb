```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
#simulo 3 porte , in una c'è una macchina, in due c'è una capra
N = np.arange(10, 10010, 500)

switcher_tot = []
conservative_tot = []
newcomer_tot = []



for j in N:
    switcher_wins = 0
    conservative_wins = 0
    newcomer_wins = 0

    for i in range(0, j):
        doors = np.zeros(3)
        doors[np.random.randint(0, 2)] = 1  

        #the switcher and conservative
        choice = random.randint(0,2)
        #probability = 1/3
        opendoor = random.randint(0,2)
        while opendoor == choice or doors[opendoor] == 1:
            opendoor = random.randint(0,2)

        secondchoice = random.randint(0,2)
        while secondchoice == opendoor or secondchoice == choice:
            secondchoice = random.randint(0,2)
        #probability = 2/3

        if doors[secondchoice] == 1:
            switcher_wins += 1
        if doors[choice] == 1:
            conservative_wins += 1

        #the newcomer
        closed_doors = doors[doors != opendoor]
        newcomer_choice = random.randint(0,1)
        #probability = 1/2
        if doors[newcomer_choice] == 1:
            newcomer_wins += 1

    switcher_tot.append(switcher_wins)
    conservative_tot.append(conservative_wins)
    newcomer_tot.append(newcomer_wins)

twothirdsline  = 2/3 * N
onehalfline = 1/2 *N 
onethirdline = 1/3 *N
plt.figure (figsize = (10, 6))
plt.plot (N, twothirdsline, color = "slategray", linestyle = "--", label = "P = 2/3")
plt.plot (N, switcher_tot, color = "slategray", label = "switcher wins")
plt.plot (N, onethirdline, color = "thistle", linestyle = "--", label = "P = 1/3")
plt.plot (N, conservative_tot, color = "thistle", label = "conservative wins")
plt.plot (N, onehalfline, color = "palegreen", linestyle = "--", label = "P = 1/2")
plt.plot (N, newcomer_tot, color = "palegreen", label = "newcomer wins")
plt.xlabel ("Number of tries")
plt.ylabel ("Number of wins")
plt.legend()
plt.tight_layout()
plt.show()
```
